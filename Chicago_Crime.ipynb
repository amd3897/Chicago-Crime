```python
import pandas as pd
fpath = 'Crimes_-_2001_to_Present.csv'
df = pd.read_csv(fpath)
df.info()
df.head()
```

```python
import os
OUTPUT_FOLDER = 'Data/Chicago/'
os.makedirs(OUTPUT_FOLDER, exist_ok=True)
```

```python
chicago_df = df
chicago_df
```

```python
date_format = "%m/%d/%Y %H:%M:%S %p"

chicago_df['Datetime'] = pd.to_datetime(chicago_df['Date'], format=date_format)
chicago_df = chicago_df.sort_values('Datetime')
chicago_df
```

```python
# Separate the Full Dataset by Years
chicago_df['Datetime'].dt.year
```

```python
# Removing 2023
filter_pre_2023 =  chicago_df['Datetime'].dt.year<2023
chicago_2023 = chicago_df[~filter_pre_2023]
chicago_df = chicago_df[filter_pre_2023]
chicago_df.head()
```

```python
# save the years for every crime
chicago_df["Year"] = chicago_df['Datetime'].dt.year.astype(str)
chicago_df["Year"].sort_values().value_counts()
```

```python
## Dropping unneeded columns to reduce file size
drop_cols = ["X Coordinate","Y Coordinate", "Community Area","FBI Code",
             "Case Number","Updated On",'Block','Location','IUCR']
```

```python
# save final df
chicago_final = chicago_df.drop(columns=drop_cols)
chicago_final = chicago_final.set_index('Datetime')
chicago_final
```

```python
# unique # of year bins
year_bins = chicago_final['Year'].astype(str).unique()
year_bins
```

```python
FINAL_DROP = ['Year']
```

```python
## set save location 

os.makedirs(OUTPUT_FOLDER, exist_ok=True)
print(f"[i] Saving .csv's to {OUTPUT_FOLDER}")
## loop through years
for year in year_bins:
    
    ## save temp slices of dfs to save.
    temp_df = chicago_final.loc[year]
    temp_df = temp_df.sort_index()
    temp_df = temp_df.reset_index(drop=True)
    temp_df = temp_df.drop(columns=FINAL_DROP)

    # save as csv to output folder
    fname_temp = f"{OUTPUT_FOLDER}Chicago-Crime_{year}.csv"#.gz
    temp_df.to_csv(fname_temp,index=False)

    print(f"- Succesfully saved {fname_temp}")
```

#### Comparing Police Districts
* Which district had the most crimes in 2022?
    * District 8 had the most with a total of 14809 cases
* Which had the least?
    * District 31 had the least with a total of 14 cases

```python
crime_22 = chicago_final.loc['2022']
crime_22.head()
```

```python
crime_districts = crime_22.groupby('District')['Primary Type']
```

```python
crime_counts = crime_districts.count().reset_index()
crime_counts.columns = ['District', 'Count']
crime_counts
```

```python
most_common = crime_counts.loc[crime_counts['Count'].idxmax()]
least_common = crime_counts.loc[crime_counts['Count'].idxmin()]
print(f'Most common:\n{most_common}\n')
print(f'Least common:\n{least_common}')
```

```python
import matplotlib.pyplot as plt

# Bar chart of total crimes by district
plt.figure(figsize=(12,6))
bars = plt.bar(crime_counts['District'], crime_counts['Count'], color='darkgreen', alpha=0.3)
plt.xlabel('Police District')
plt.ylabel('Total Crimes')
plt.title('Total Crimes by Police District in Chicago')
plt.xticks(crime_counts['District'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 100, round(yval),
             ha='center', va='top', rotation=270, fontsize=12)
    
plt.tight_layout()
plt.show()
```

#### Crimes Across the Years:
* Is the total number of crimes increasing or decreasing across the years?
    * The total number of crimes is decreasing.
    
* Are there any individual crimes that are doing the opposite (e.g., decreasing when overall crime is increasing or vice-versa)?
    * Generally, all individual crimes are decreasing.

```python
# Is the total number of crimes increasing or decreasing across the years?
crime_by_year = chicago_final.groupby('Year')['Primary Type'].count()
crime_by_year.sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,6))

plt.plot(crime_by_year.index, crime_by_year.values, marker='o', color='darkgreen')
plt.title('Number of Crimes by Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Crimes', fontsize=14)
plt.xticks(ha='center', rotation=90)

plt.axvline(crime_by_year.idxmax(), color='green', linestyle='--', linewidth=2)
plt.axvline(crime_by_year.idxmin(), color='red', linestyle='--', linewidth=2)

plt.legend(['Number of Crimes', 'Max', 'Min'], fontsize=12, loc='upper center')
plt.tight_layout()
plt.show();
```

```python
# Are there any individual crimes that are doing the opposite (e.g., decreasing when overall crime is increasing or vice-versa)?
individual_crime = chicago_final.groupby(['Year', 'Primary Type']).size().unstack(fill_value=0)
individual_crime.head()
```

```python
#fig, axes = plt.subplots(len(crime_type_groups), 1, figsize=(12, 4 * len(crime_type_groups)), sharex=True)
individual_crime.plot(figsize=(15, 10))
plt.subplots_adjust()
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Crimes', fontsize=14)
plt.title('Trends of Individual Crimes Over Time in Chicago', fontsize=16)
plt.legend(title='Crime Type', bbox_to_anchor=(1, 1), loc='upper left', fontsize=12)
plt.tight_layout()
plt.show()
```

#### Comparing Months:
* What months have the most crime? What months have the least?
    * July has the most with a total of 717267 cases. And February has the least with a total of 529417 cases.
* Are there any individual crimes that do not follow this pattern? If so, which crimes?

```python
crime_by_month = chicago_final.reset_index()
crime_by_month.head()
```

```python
crime_by_month['Month'] = crime_by_month['Datetime'].dt.month
crime_by_month. head()
```

```python
crime_by_month = crime_by_month.drop(columns='Year')
crime_by_month = crime_by_month.set_index('Datetime')
crime_by_month.head()
```

```python
# What months have the most crime? What months have the least?
total_crime = crime_by_month.groupby('Month')['Primary Type'].count()
total_crime
```

```python
plt.figure(figsize=(10, 6))
plt.bar(total_crime.index, total_crime.values, color='darkgreen', alpha=0.3)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total Crimes', fontsize=14)
plt.title('Total Crimes by Month in Chicago', fontsize=16)
plt.xticks(total_crime.index)

max_month = total_crime.idxmax()
max_value = total_crime.max()
min_month = total_crime.idxmin()
min_value = total_crime.min()
plt.annotate(f'Top Month: {max_month}\n Total crimes: {max_value}', xy=(max_month, max_value), 
             ha='center', va='top', fontsize=12)
plt.annotate(f'Least Month: {min_month}\n Total crimes: {min_value}', xy=(min_month, min_value),
             ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
individual_crimes_by_month = crime_by_month.groupby(['Month', 'Primary Type']).size().unstack(fill_value=0)

individual_crimes_by_month.plot(figsize=(12, 8), marker='o')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.title('Trends of Individual Crimes Over Months in Chicago')

plt.legend(title='Crime Type', bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

```python
crime_by_month['Primary Type'].unique()
```

```python
crime_type_groups = {
    'Group 1': ['OTHER OFFENCE', 'THEFT','BURGLARY', 'CRIMINAL DAMAGE', 'BATTERY','MOTOR VEHICLE THEFT'],
    'Group 2': ['ASSAULT','NARCOTICS','OFFENCE INVOLVING CHILDREN','PUBLIC PEACE VIOLATION','WEAPONS VIOLATION','CRIM SEXUAL ASSAULT'],
    'Group 3': ['GAMBLING','SEX OFFENCE','DECEPTIVE PRACTICE','LIQUOR LAW VIOLATION','ROBBERY','CRIMINAL TRESPASS'],
    'Group 4': ['INTERFERANCE WITH PUBLIC OFFICER','HOMICIDE','KIDNAPPING','PROSTITUTION','STALKING','INTIMIDATION'],
    'Group 5': ['CRIMINAL SEXUAL ASSAULT','ARSON','RITUALISM','DOMESTIC VIOLENCE','OBSCENITY','OTHER NARCOTIC VIOLATION'],
    'Group 6': ['PUBLIC INDECENCY','CONCEALED CARRY LICENSE VIOLATION','NON-CRIMINAL','HUMAN TRAFFICKING','NON-CRIMINAL (SUBJECT SPECIFIED)','NON - CRIMINAL'],
}
crime_by_month['crime_group'] = crime_by_month['Primary Type'].apply(lambda x: next((group for group, types in crime_type_groups.items() if x in types), 'Other'))
crime_by_month.head()
```

```python
fig, axes = plt.subplots(len(crime_type_groups), 1, figsize=(12, 4 * len(crime_type_groups)), sharex=True)

for i, group in enumerate(crime_type_groups):
    # Filter data for the current group
    group_data = crime_by_month[crime_by_month['crime_group'] == group]
    
    # Group by month and crime type, count the number of crimes, and reshape the data
    crimes_by_group = group_data.groupby(['Month', 'Primary Type']).size().unstack(fill_value=0)
    
    # Plot the data on the current subplot
    crimes_by_group.plot(ax=axes[i], marker='o', legend=False)
    
    # Set labels and title for the subplot
    axes[i].set_xlabel('Month')
    axes[i].set_ylabel('Number of Crimes')
    axes[i].set_title(f'Trends of Crimes in Group: {group}')
    
plt.xticks(crimes_by_group.index)
plt.legend(title='Crime Type', bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
