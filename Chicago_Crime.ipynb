## Part 1

```python
import pandas as pd
fpath = 'Crimes_-_2001_to_Present.csv'
df = pd.read_csv(fpath)
df.info()
df.head()
```

```python
import os
OUTPUT_FOLDER = 'Data/Chicago/'
os.makedirs(OUTPUT_FOLDER, exist_ok=True)
```

```python
chicago_df = df
chicago_df.head()
```

```python
date_format = "%m/%d/%Y %H:%M:%S %p"

chicago_df['Datetime'] = pd.to_datetime(chicago_df['Date'], format=date_format)
chicago_df = chicago_df.sort_values('Datetime')
chicago_df.head()
```

```python
# Separate the Full Dataset by Years
chicago_df['Datetime'].dt.year
```

```python
# Removing 2023
filter_pre_2023 =  chicago_df['Datetime'].dt.year<2023
chicago_2023 = chicago_df[~filter_pre_2023]
chicago_df = chicago_df[filter_pre_2023]
chicago_df.head()
```

```python
# save the years for every crime
chicago_df["Year"] = chicago_df['Datetime'].dt.year.astype(str)
chicago_df["Year"].sort_values().value_counts()
```

```python
## Dropping unneeded columns to reduce file size
drop_cols = ["X Coordinate","Y Coordinate", "Community Area","FBI Code",
             "Case Number","Updated On",'Block','Location','IUCR']
```

```python
# save final df
chicago_final = chicago_df.drop(columns=drop_cols)
chicago_final = chicago_final.set_index('Datetime')
chicago_final.head()
```

```python
# unique # of year bins
year_bins = chicago_final['Year'].astype(str).unique()
year_bins
```

```python
FINAL_DROP = ['Year']
```

```python
## set save location 

os.makedirs(OUTPUT_FOLDER, exist_ok=True)
print(f"[i] Saving .csv's to {OUTPUT_FOLDER}")
## loop through years
for year in year_bins:
    
    ## save temp slices of dfs to save.
    temp_df = chicago_final.loc[year]
    temp_df = temp_df.sort_index()
    temp_df = temp_df.reset_index(drop=True)
    temp_df = temp_df.drop(columns=FINAL_DROP)

    # save as csv to output folder
    fname_temp = f"{OUTPUT_FOLDER}Chicago-Crime_{year}.csv"#.gz
    temp_df.to_csv(fname_temp,index=False)

    print(f"- Succesfully saved {fname_temp}")
```

```python
def plot_forecast(ts_train, ts_test, forecast_df, n_train_lags=None, 
                  figsize=(10,4), title='Comparing Forecast vs. True Data'):
    ### PLot training data, and forecast (with upper/,lower ci)
    fig, ax = plt.subplots(figsize=figsize)

    # setting the number of train lags to plot if not specified
    if n_train_lags==None:
        n_train_lags = len(ts_train)
            
    # Plotting Training  and test data
    ts_train.iloc[-n_train_lags:].plot(ax=ax, label="train")
    ts_test.plot(label="test", ax=ax)

    # Plot forecast
    forecast_df['mean'].plot(ax=ax, color='green', label="forecast")

    # Add the shaded confidence interval
    ax.fill_between(forecast_df.index, 
                    forecast_df['mean_ci_lower'],
                   forecast_df['mean_ci_upper'],
                   color='green', alpha=0.3,  lw=2)

    # set the title and add legend
    ax.set_title(title)
    ax.legend();
    
    return fig, ax
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error

def regression_metrics_ts(ts_true, ts_pred, label="", verbose=True, output_dict=False,):
    # Get metrics
    mae = mean_absolute_error(ts_true, ts_pred)
    mse = mean_squared_error(ts_true, ts_pred)
    rmse = mean_squared_error(ts_true, ts_pred, squared=False)
    r_squared = r2_score(ts_true, ts_pred)
    mae_perc = mean_absolute_percentage_error(ts_true, ts_pred) * 100

    if verbose == True:
        # Print Result with label
        header = "---" * 20
        print(header, f"Regression Metrics: {label}", header, sep="\n")
        print(f"- MAE = {mae:,.3f}")
        print(f"- MSE = {mse:,.3f}")
        print(f"- RMSE = {rmse:,.3f}")
        print(f"- R^2 = {r_squared:,.3f}")
        print(f"- MAPE = {mae_perc:,.2f}%")

    if output_dict == True:
        metrics = {
            "Label": label,
            "MAE": mae,
            "MSE": mse,
            "RMSE": rmse,
            "R^2": r_squared,
            "MAPE(%)": mae_perc,
        }
        return metrics
```

```python
# Custom function for Ad Fuller Test
def get_adfuller_results(ts, alpha=.05, label='adfuller', **kwargs): #kwargs for adfuller()
    # Saving each output
    (test_stat, pval, nlags, nobs, crit_vals_d, 
    icbest ) = tsa.adfuller(ts, **kwargs)
    # Converting output to a dictionary with the interpretation of p
    adfuller_results = {'Test Statistic': test_stat,
                        "# of Lags Used":nlags, 
                       '# of Observations':nobs,
                        'p-value': round(pval,6),
                        'alpha': alpha,
                       'sig/stationary?': pval < alpha}
    return pd.DataFrame(adfuller_results, index =[label])
```

```python
def plot_acf_pacf(ts, nlags=40, figsize=(10, 5), 
                  annotate_sig=False, alpha=.05,
                 acf_kws={}, pacf_kws={},  
                  annotate_seas=False, m = None,
                 seas_color='black'):
    
    fig, axes = plt.subplots(nrows=2, figsize=figsize)

    
    # Sig lags line style
    sig_vline_kwargs = dict( ls=':', lw=1, zorder=0, color='red')

    # ACF
    tsa.graphics.plot_acf(ts, ax=axes[0], lags=nlags, **acf_kws)
    
    ## Annotating sig acf lags
    if annotate_sig == True:
        sig_acf_lags = get_sig_lags(ts,nlags=nlags,alpha=alpha, type='ACF')
        for lag in sig_acf_lags:
            axes[0].axvline(lag,label='sig', **sig_vline_kwargs )

    # PACF
    tsa.graphics.plot_pacf(ts,ax=axes[1], lags=nlags, **pacf_kws)
    
    ## Annotating sig pacf lags
    if annotate_sig == True:
        ## ANNOTATING SIG LAGS
        sig_pacf_lags = get_sig_lags(ts,nlags=nlags,alpha=alpha, type='PACF')
        for lag in sig_pacf_lags:
            axes[1].axvline(lag, label='sig', **sig_vline_kwargs)



    
    ### ANNOTATE SEASONS
    if annotate_seas == True:
        # Ensure m was defined
        if m is None:
            raise Exception("Must define value of m if annotate_seas=True.")

        ## Calculate number of complete seasons to annotate
        n_seasons = nlags//m 

        # Seasonal Lines style
        seas_vline_kwargs = dict( ls='--',lw=1, alpha=.7, color=seas_color, zorder=-1)
        
        ## for each season, add a line
        for i in range(1, n_seasons+1):
            axes[0].axvline(m*i, **seas_vline_kwargs, label="season")
            axes[1].axvline(m*i, **seas_vline_kwargs, label="season")

    fig.tight_layout()
    
    return fig
```

```python
# Decompose the Time series to detect seasonality and decide whether to use a seasonal model or not
import statsmodels.tsa.api as tsa
from pmdarima.model_selection import train_test_split
from pmdarima.arima.utils import ndiffs, nsdiffs

decomp_theft = tsa.seasonal_decompose(ts_theft)
fig = decomp_theft.plot()
fig.set_size_inches(12,5)
fig.tight_layout()
```

```python
seasonal = decomp_theft.seasonal
ax = seasonal.loc['2016':'2019'].plot(marker = 'o') # seasonal component m = 12
```

```python
get_adfuller_results(ts_theft) # non-stationary
```

```python
# Determine if nonseasonal and/or seasonal differencing is required
d = ndiffs(ts_theft)
D = nsdiffs(ts_theft, m=12)
print(f"d = {d}, D = {D}")
```

```python
ts_theft_diff = ts_theft.diff().dropna()
```

```python
get_adfuller_results(ts_theft_diff) # stationarity
```

```python
# Use the ACF and PACF plots of stationarity data to estimate initial orders
plot_acf_pacf(ts_theft_diff, annotate_seas=True, m = 12);
# ACF gradually decrease and PACF has dramatic decrease for seasonality P = 1, D = 0, Q = 1
# ACF and PACF gradually decrease for non seasonality p = 1, d = 1, q = 1
```

```python
# Split the time series into training and test data (Remember we want to predict 6 months)
train, test = train_test_split(ts_theft, test_size=6)
ax = train.plot(label='Train')
test.plot(ax=ax, label='Test', figsize=(10, 6))
ax.legend();
```

```python
# Fit a manual ARIMA/SARIMA model based on the orders determined during your exploration.
sarima = tsa.ARIMA(train, order = (1, 1, 1), seasonal_order = (1, 0, 1, 12)).fit()
```

```python
# Obtain summary of forecast as dataframe
forecast_df = sarima.get_forecast(len(test)).summary_frame()
# Plot the forecast with true values
plot_forecast(train, test, forecast_df)
# Obtain metrics
regression_metrics_ts(test, forecast_df["mean"])
```

```python
sarima.summary()
```

```python
# Obtain diagnostic plots
fig = sarima.plot_diagnostics()
fig.set_size_inches(10,6)
fig.tight_layout()
```

```python
# Tune with pmdarima's auto_arima
import pmdarima as pm
auto_model = pm.auto_arima(train, seasonal=True, m=12, trace=True, stepwise=True)
```

```python
auto_model.summary()
```

```python
fig = auto_model.plot_diagnostics()
fig.set_size_inches(10,6)
fig.tight_layout()
```

```python
# Make forecasts with the auto_arima model
auto_model = tsa.ARIMA(train, order=auto_model.order, seasonal_order=auto_model.seasonal_order).fit()
auto_forecast = auto_model.get_forecast(steps=len(test)).summary_frame()
# Plot the forecasts versus the test data
plot_forecast(train, test, auto_forecast)
# Obtain metrics for evaluation
regression_metrics_ts(test, auto_forecast['mean'])
```

* The auto model is better than the manual model
* R^2 = 0.613 (auto model) vs R^2 = 0.513 (manual model)
* MAPE = 4% (auto model) vs MAPE = 4.91% (manual model)
* RMSE = 260.955 (auto model) vs RMSE = 292.783 (manual model)

```python
# Make true future forecasts using the entire dataset (for dates beyond the entire ts)
future_forecast = auto_model.get_forecast(steps=6)
forecast_final_value = future_forecast.predicted_mean[0]
```

```python
future_dates = pd.date_range(start=test.index[-1], periods=7, freq='M')
predictions = auto_model.predict(start=future_dates[1], end=future_dates[-1])
print(predictions.round(0))
```
